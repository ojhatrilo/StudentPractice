```python
import pandas as pd
df = pd.read_excel("Customer Call List.xlsx")
df
```

```python
df = df.drop(columns="Not_Useful_Column")
df
```

```python
df.drop_duplicates(inplace=True)
df
```

```python
df.dropna(inplace=True)
df
```

```python
df["Last_Name"] = df['Last_Name'].str.replace("_","")
df
```

```python
df["Phone_Number"] = df["Phone_Number"].str.replace('/',"")
df["Phone_Number"] = df["Phone_Number"].str.replace('|',"")
df["Phone_Number"] = df["Phone_Number"].str.replace('-',"")
df
```

```python
df["Phone_Number"] 
```

```python
df.drop(index=14)
df
```

```python
for x in df.index:
    if df.loc[x, "Phone_Number"] == 'Na':
        df.drop(x, inplace=True)

df
```

```python
for x in df.index:
    if df.loc[x, 'Do_Not_Contact'] == "Y" or df.loc[x, 'Do_Not_Contact'] == "Yes":
        df.drop(x, inplace=True)

df
```

```python
df1 = df.loc[df["Do_Not_Contact"] == "N" ]
df1["Do_Not_Contact"] = df["Do_Not_Contact"].str.replace("N","No")
df1
```

```python
df['Do_Not_Contact'] = df["Do_Not_Contact"].str.replace("N","No")
df
```

```python
df['Do_Not_Contact'] = df["Do_Not_Contact"].str.replace("Noo","No")
df
```

```python
df["Paying Customer"] = df["Paying Customer"].str.replace("Y","Yes")
df
```

```python
df["Paying Customer"] = df["Paying Customer"].str.replace("Yeses","Yes")
df
```

```python
df['Paying Customer'] = df["Paying Customer"].str.replace("N","No")
df['Paying Customer'] = df["Paying Customer"].str.replace("Noo","No")
df
```

```python
df = df.reset_index(drop=True)
df
```

```python
df.to_excel("Cleaning2.xlsx")
```
